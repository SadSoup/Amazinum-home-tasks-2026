##  Brest cancer  dataset
</font>



```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import tree
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names
print ('labels:',labels)
print ('features:', features)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

<font color = green>

##  DecisionTree 

</font>



```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = DecisionTreeClassifier(
    criterion=  'entropy', # gini showed worse results
    random_state=0,
    max_depth=3,
    # experimenting with min_samples_leaf and min_samples_splitdid not show better results
    ).fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
graph_viz = tree.export_graphviz(clf, out_file=None,feature_names=features, class_names=labels, filled = True)
graph = graphviz.Source(graph_viz)
graph.view(cleanup =True) # cleanup (bool) – Delete the source file after rendering.
```

<font color = green>

##  RandomForest

</font>



```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = RandomForestClassifier(
    n_estimators=50,
    max_depth=5,
    min_samples_leaf=4,
    max_features= 10,
    random_state= 0
).fit(X_train, y_train) 
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```

<font color = green>

##  Gradient Boosting Decision Trees (GBDT)

</font>



```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = GradientBoostingClassifier( 
    n_estimators=200,
    subsample=0.8,
    learning_rate=0.1,
    max_depth=2,
    min_samples_leaf=4,
    random_state= 0
).fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```
